```python
m1 = np.zeros((h+2,w+2))
m2 = np.zeros((h+2,w+2)) 
m3 = np.zeros((h+2,w+2))
m4 = np.zeros((h+2,w+2)) 
m5 = np.zeros((h+2,w+2))


#copy m2
for i in range(h):
    for j in range(w):
        m2[i+2][j+2] = np.int64(img[i][j])
        
for i in range(h):
    for j in range(w):
        m1[i][j] = img[i-(i%2)][j-(j%2)]
        m2[i][j] = img[i-(i%4)][j-(j%4)]
        m3[i][j] = img[i-(i%8)][j-(j%8)]
        m4[i][j] = img[i-(i%16)][j-(j%16)]
        m5[i][j] = img[i-(i%32)][j-(j%32)]
    
    
plt.figure(1)
plt.title('Original Image')
plt.xlabel('Pixel Position X')
plt.ylabel('Pixel Position Y')
plt.imshow(img,cmap="gray")
plt.figure(2)
plt.title('After Gaussian (Scale 1/2)')
plt.xlabel('Pixel Position X')
plt.ylabel('Pixel Position Y')
plt.imshow(m1,cmap="gray")
plt.figure(3)
plt.title('After Gaussian (Scale 1/4)')
plt.xlabel('Pixel Position X')
plt.ylabel('Pixel Position Y')
plt.imshow(m2,cmap="gray")
plt.figure(4)
plt.title('After Gaussian (Scale 1/8)')
plt.xlabel('Pixel Position X')
plt.ylabel('Pixel Position Y')
plt.imshow(m3,cmap="gray")
plt.figure(5)
plt.title('After Gaussian (Scale 1/16)')
plt.xlabel('Pixel Position X')
plt.ylabel('Pixel Position Y')
plt.imshow(m4,cmap="gray")
plt.figure(6)
plt.title('After Gaussian (Scale 1/32)')
plt.xlabel('Pixel Position X')
plt.ylabel('Pixel Position Y')
plt.imshow(m5,cmap="gray")
```

```python

def applyGauss(img,h,w):
    m2 = np.zeros((h+2,w+2))
    m3 = np.zeros((h+2,w+2)) 

    #copy m2
    for i in range(h):
        for j in range(w):
            m2[i+2][j+2] = np.int64(img[i][j])


    for i in range(h-2):
        for j in range(w-2):
            tl = np.uint32(img[i][j])
            tr = np.uint32(img[i][j+1])
            bl = np.uint32(img[i+1][j])
            br = np.uint32(img[i+1][j+1])
            res = (tl+tr+bl+br)
            m3[i][j] = res/16
    #         if res>max:
    #             max = res
    

    plt.figure(1)
    plt.imshow(img,cmap="gray")
    plt.figure(2)
    plt.imshow(m3,cmap="gray")
    return m3
    
#img = applyGauss(applyGauss(img,h,w) ,h,w) 
applyGauss(img,h,w) 
```
